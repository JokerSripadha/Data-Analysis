• Taking a dataset of 26 states with features like Literacy, Cleanliness,

Crime Rate and targetting (predicting) Good or Bad state!

• Good is called Target variable here, it has values of Os and 1s



Now  Let Us  create  a decision tree
- Decision tree recurrently spilts the data until it gets pure leaves.
- Let us view a DT based on Crime.

building decision tree:-
- NOde 1:  CR>60
- True-[C,Q,X]=O(pure leaf)
- FAlse-[A,E,F,G,I,K,L,N,P,R,U,V=0][B,D,H,J,M,O,S,T,W,Y,Z=1]
the mixed leaf has target variable as both 0s and 1s. hence this data is spilted once again

NODE 2: CR>50
- True-[A,E,F,G,I,K,L,N,P,R,U,V=0](pure leaf)
- False-[B,D,H,J,M,O,S,T,W,Y,Z=1](pure leaf)



```python
import pandas as pd
```

```python
df=pd.read_csv("/content/demodt.txt",sep=',')
```

```python
df
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
target = df.Good
target
```

```python
feat_list=['Literacy','Cleanliness','Crime_Rate']
feat=df[feat_list]

feat
```
